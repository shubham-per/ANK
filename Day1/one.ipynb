```python
import numpy as np
import pandas as pd
```

```python
tdf=pd.read_csv("temperature_data.csv")
tdf.head(10)
```

```python
print(tdf.isnull().sum())    #checking missing value
```

```python
#tdf['Temperature'] = tdf['Temperature'].fillna(tdf['Temperature'].mean())
```

```python
#Since the temperature appears to increase steadily, linear interpolation would be a good choice
tdf['Temperature'] = tdf['Temperature'].interpolate()
```

```python
tdf['Condition'] = tdf['Condition'].astype('category').cat.codes
```

```python
tdf.to_csv('filled_data.csv', index=False)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(tdf['Time'], tdf['Temperature'], c='r', lw=2)
plt.xlabel('Time (hours)')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Trends Over Time(lineplot)')
plt.grid()
plt.show()
```

```python
plt.scatter(tdf['Time'], tdf['Temperature'], marker="^", s=2, alpha=0.4)
plt.xlabel('Time (hours)')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Distribution(scatterplot)')
plt.grid()
plt.show()
```

```python
plt.hist(tdf['Temperature'], bins=30)
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.title('Temperature Distribution')
plt.show()
```
